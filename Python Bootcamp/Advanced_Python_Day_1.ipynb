```python
#DataTypes
a=2
b=2.2
print(type(a))
print(type(b))
```

```python
#Strings
c="Ed Spread"
print(type(c))
```

```python
#Lists
d=[21,32,534,77,665,"Python"]
print(type(d))
#length
print(len(d))
#indexing
print(d[3])
```

```python
#Tuples
e=(12,32,54,43,"Python")
print(type(e))
print(len(e))
#Concatinate
f=(1,43,56,54,"Scope")
e+f
#Indexing
print(f[4])
```

```python
#Sets
a={"Ghost",12,23,32,54,32}
print(type(a))
#Union
b={23,43,"Set",56,87,54}
print(a|b)
#Intersection
print(a&b)
```

```python
#Dictionary
a={1:32,2:66,3:"Onslaught"}
print(type(a))
print(len(a))
```

```python
#Arithmetic Operators
#Addition
print(50+50)
#Substraction
print(50-50)
#Multiplication
print(50*50)
#Division
print(50/50)
#No Remainder
print(50//6)
#Exponents
print(50**2)
#Modulus
print(50%6)
```

```python
#Logical Operations
a=[12,43,564,23,54,75,"Python"]
print(43 in a)
print("Python" not in a)
```
