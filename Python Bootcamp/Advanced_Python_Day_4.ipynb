```python
class Student:
  def __init__(self,rollno,name,fees,marks,grade):
    self.rollno=rollno
    self.name=name
    self.fees=fees
    self.marks=marks
    self.grade=grade
class Fees(Student):
  def recieptgen(self):
    print("The fees to be payed: ",self.fees)
class Result(Student):
  def result(self):
    print("Marks obtained by the student: ",self.marks)
    print("Grade obtained by the student: ",self.grade)
rollno=int(input("Enter the roll no.: "))
name=input("Enter the name: ")
fees=float(input("Enter the fees required: "))
marks=int(input("Enter the marks recieved: "))
grade=input("Enter the grade recieved: ")
obj1=Fees(rollno,name,fees,marks,grade)
obj2=Result(rollno,name,fees,marks,grade)
print(obj1.recieptgen())
print(obj2.result())
```

```python
class Distance:
  def __init__(self,OM,OKM,SKM,SM):
    self.OM=OM
    self.OKM=OKM
    self.SKM=SKM
    self.SM=SM
class School(Distance):
  def dispschool(self):
    print("School distance in metres: ",self.SM)
    print("School distance in kilometres: ",self.SKM)
class Office(Distance):
  def dispoffice(self):
    print("Office distance in metres: ",self.OM)
    print("Office distance in kilometres ",self.OKM)
OM=float(input("Enter the Office distance in metres: "))
OKM=float(input("Enter the Office distance in kilometres: "))
SM=float(input("Enter the School distance in metres: "))
SKM=float(input("Enter the School distance in kilometres: "))
obj1=School(OM,OKM,SKM,SM)
obj2=Office(OM,OKM,SKM,SM)
print(obj1.dispschool())
print(obj2.dispoffice())
```

```python
class Train:
  def __init__(self,tier1,tier2,tier3,btier1,btier2,btier3,tstatus):
    self.tier1=tier1
    self.tier2=tier2
    self.tier3=tier3
    self.btier1=btier1
    self.btier2=btier2
    self.btier3=btier3
    self.tstatus=tstatus
  def disp(self):
    print("Number of Tier 1 seats: ",self.tier1)
    print("Number of Tier 2 seats: ",self.tier2)
    print("Number of Tier 3 seats: ",self.tier3)
    print("Number of Booked Tier 1 seats: ",self.btier1)
    print("Number of Booked Tier 2 seats: ",self.btier2)
    print("Number of Booked Tier 3 seats: ",self.btier3)
class Reservation(Train):
  def treserve(self):
    if tstatus=="Booked":
      print("Ticked reservation done")
    elif tstatus=="Cancelled":
      print("Ticket is cancelled")
    else:
      return "Ticket not Booked"
tier1=int(input("Enter the number of Tier 1 seats: "))
tier2=int(input("Enter the number of Tier 2 seats: "))
tier3=int(input("Enter the number of Tier 3 seats: "))
btier1=int(input("Enter the number of Booked Tier 1 seats: "))
btier2=int(input("Enter the number of Booked Tier 2 seats: "))
btier3=int(input("Enter the number of Booked Tier 3 seats: "))
tstatus=input("Enter ticked status: ")
obj1=Train(tier1,tier2,tier3,btier1,btier2,btier3,tstatus)
obj2=Reservation(tier1,tier2,tier3,btier1,btier2,btier3,tstatus)
print(obj1.disp())
print(obj2.treserve())
```
