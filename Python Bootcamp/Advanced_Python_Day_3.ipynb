```python
#Class Syntax
class calculator:
  def __init__(self,a,b):
    self.a=a
    self.b=b
  def add(self,a,b):
    print("Addition is: ", self.a+self.b)
  def sub(self,a,b):
    print("Substraction is: ",self.a-self.b)
  def mult(self,a,b):
    print("Multiplication is: ",self.a*self.b)
  def div(self,a,b):
    print("Division is: ",self.a/self.b)
a=int(input("Enter the first number: "))
b=int(input("Enter the second number: "))
obj=calculator(a,b)
print(obj.add(a,b))
print(obj.sub(a,b))
print(obj.mult(a,b))
print(obj.div(a,b))
```

```python
#Class Rectangle
class rectangle:
  def __init__(self,length,breadth):
    self.length=length
    self.breadth=breadth
  def perimeter(self):
    return 2*(self.length+self.breadth)
  def area(self):
    return self.length*self.breadth
length=int(input("Enter the length of the rectangle: "))
breadth=int(input("Enter the breadth of the rectangle: "))
dimensions=rectangle(length,breadth)
print(dimensions.perimeter())
print(dimensions.area())
```

```python
#Class Book
class book:
  def __init__(self,Title,Author,Price):
    self .Title=Title
    self.Author=Author
    self.Price=Price
  def view(self):
    print("Title of the Book: ",self.Title)
    print("Author: ",self.Author)
    print("Price: ",self.Price)
Title=input("Enter the name of the Book: ")
Author=input("Enter the name of the Author: ")
Price=float(input("Enter the Price of the Book: "))
obj=book(Title,Author,Price)
print(obj.view())
```
