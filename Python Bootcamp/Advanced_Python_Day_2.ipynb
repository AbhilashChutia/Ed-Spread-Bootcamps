# Loops and Functions

```python
#For loop
#Print 10 numbers
for i in range(1,11):
  print(i)
```

```python
#While Loop
a=int(input("Enter a number: "))
while a>5:
  print("The number is acceptable!")
  break
```

```python
#Examples of For loop
name = input("Enter a name: ")
for u in range(1,5):
  print(name)
```

```python
#Example of While loop
i=1
while i<5:
  print(i)
  i=i+1
```

```python
#If Statement
#Major or Minor check
age=int(input("Enter the age: "))
if age>18:
  print("Major")
else:
  print("Minor")
```

```python
#Example of If Statement
a=int(input("Enter your number A: "))
b=int(input("Enter your number B: "))
c=int(input("Enter your number C: "))
if a>b and a>c:
  print("A is the largest number")
elif b>a and b>c:
  print("B is the largest number")
elif c>a and c>b:
  print("C is the largest number")
else:
  print("Invalid Input!!")
```

```python
#Functions
def addition(a,b):
  print("The sum of is: ", a+b )
addition(7,5)
```

```python
#Examples of Functions
s=input("Enter your sex:")
def scheck(s):
  if s=="M":
    print("Sex is Male")
  elif s=="F":
    print("Sex is Female")
  else:
    print("Invalid Input")
scheck(s)
```

```python
#Lambda Function
x=lambda a,b:a+b
print(x(5,10))
#Function
def add(a,b):
  print(a+b)
add(5,10)
```
